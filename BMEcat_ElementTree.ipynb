```python

catalog_structure_node = ET.SubElement(catalog_group_system, 'CATALOG_STRUCTURE')
catalog_structure_node.set('type', "node")

group_id = ET.SubElement(catalog_structure_node, 'GROUP_ID')
group_id.text = '2'

group_name = ET.SubElement(catalog_structure_node, 'GROUP_NAME')
group_name.text = '{} Products'.format(str(input('Company short name: ')))

group_description = ET.SubElement(catalog_structure_node, 'GROUP_DESCRIPTION')
group_description.text = group_name.text

parent_id = ET.SubElement(catalog_structure_node, 'PARENT_ID')
parent_id.text ='1'

group_order = ET.SubElement(catalog_structure_node, 'GROUP_ORDER')
group_order.text = '2'
```

```python
cg_file = pd.read_excel(str(input('Paste groupID data file: ')).replace('"', ''))
```

```python
cgID = cg_file[[str(input('Column to group and derive Catalog Group IDs: ')), 'CATALOG_GROUP_ID']].drop_duplicates(subset=['CATALOG_GROUP_ID']).reset_index(drop=True)
```

```python
i = 3
for key, value in groupID.items():
    catalog_structure = ET.SubElement(catalog_group_system, 'CATALOG_STRUCTURE')
    catalog_structure.set('type', "leaf")

    group_id = ET.SubElement(catalog_structure, 'GROUP_ID')
    group_id.text = '{key}'.format(key=key)

    group_name = ET.SubElement(catalog_structure, 'GROUP_NAME')
    group_name.text = '{value}'.format(value=value)

    group_description = ET.SubElement(catalog_structure, 'GROUP_DESCRIPTION')
    group_description.text = '{value}'.format(value=value)

    parent_id = ET.SubElement(catalog_structure, 'PARENT_ID')
    parent_id.text = '2'

    group_order = ET.SubElement(catalog_structure, 'GROUP_ORDER')
    group_order.text = '{}'.format(i)
    i = i + 1
    
```

```python

catalog_structure_root = ET.SubElement(catalog_group_system, 'CATALOG_STRUCTURE')
catalog_structure_root.set('type', "root")

group_id = ET.SubElement(catalog_structure_root, 'GROUP_ID')
group_id.text = '1'

group_name = ET.SubElement(catalog_structure_root, 'GROUP_NAME')
group_name.text = '{} Catalog'.format(str(input('Company name short: ')))

group_description = ET.SubElement(catalog_structure_root, 'GROUP_DESCRIPTION')
group_description.text = group_name.text

parent_id = ET.SubElement(catalog_structure_root, 'PARENT_ID')
parent_id.text = '0'

group_order = ET.SubElement(catalog_structure_root, 'GROUP_ORDER')
group_order.text = '1'
```

```python
cwf_dict = {}
```

```python
for index, row in cwf.iterrows():
    cwf_dict[index] = row.to_dict()
```

```python
for key, i_key in cwf_dict.items():

    article = ET.SubElement(t_new_catalog, 'ARTICLE')

    supplier_aid = ET.SubElement(article, 'SUPPLIER_AID')
    supplier_aid.text = '{}'.format(key)


    # article details
    article_details = ET.SubElement(article, 'ARTICLE_DETAILS')

    description_short = ET.SubElement(article_details, 'DESCRIPTION_SHORT')
    description_short.text = '{}'.format(i_key['DESCRIPTION_SHORT (80)'])

    description_long = ET.SubElement(article_details, 'DESCRIPTION_LONG')
    description_long.text = '{}'.format(i_key['DESCRIPTION_LONG'])

    manufacturer_name = ET.SubElement(article_details, 'MANUFACTURER_NAME')
    manufacturer_name.text = '{}'.format(i_key['MANUFACTURER_NAME'])
    
    delivery_time = ET.SubElement(article_details, 'DELIVERY_TIME')
    delivery_time.text = '{}'.format(i_key['DELIVERY_TIME'])

    keyword = ET.SubElement(article_details, 'KEYWORD')
    keyword.text = '{}'.format(i_key['KEYWORD (max 50)'])


    # article features ECLASS
    article_features = ET.SubElement(article, 'ARTICLE_FEATURES')

    reference_feature_system_name = ET.SubElement(article_features, 'REFERENCE_FEATURE_SYSTEM_NAME')
    reference_feature_system_name.text = 'ECLASS-5.1'

    reference_feature_group_id = ET.SubElement(article_features, 'REFERENCE_FEATURE_GROUP_ID')
    reference_feature_group_id.text = '{}'.format(i_key['FVALUE_5.1'])

    feature = ET.SubElement(article_features, 'FEATURE')
    fname = ET.SubElement(feature, 'FNAME')
    fname.text = 'ECLASS-5.1'
    fvalue = ET.SubElement(feature, 'FVALUE')
    fvalue.text = '{}'.format(i_key['FVALUE_5.1'])


    # article features ECCN
    article_features = ET.SubElement(article, 'ARTICLE_FEATURES')

    reference_feature_system_name = ET.SubElement(article_features, 'REFERENCE_FEATURE_SYSTEM_NAME')
    reference_feature_system_name.text = 'Export Classification Number (ECCN)'

    reference_feature_group_id = ET.SubElement(article_features, 'REFERENCE_FEATURE_GROUP_ID')
    reference_feature_group_id.text = '{}'.format(i_key['FVALUE(ECCN)'])

    feature = ET.SubElement(article_features, 'FEATURE')
    fname = ET.SubElement(feature, 'FNAME')
    fname.text = 'Export Classification Number (ECCN)'
    fvalue = ET.SubElement(feature, 'FVALUE')
    fvalue.text = '{}'.format(i_key['FVALUE(ECCN)'])

    # article features Commodity Code
    article_features = ET.SubElement(article, 'ARTICLE_FEATURES')

    reference_feature_system_name = ET.SubElement(article_features, 'REFERENCE_FEATURE_SYSTEM_NAME')
    reference_feature_system_name.text = 'Commodity Code'

    reference_feature_group_id = ET.SubElement(article_features, 'REFERENCE_FEATURE_GROUP_ID')
    reference_feature_group_id.text = '{}'.format(i_key['FVALUE_CommodityCode'])

    feature = ET.SubElement(article_features, 'FEATURE')
    fname = ET.SubElement(feature, 'FNAME')
    fname.text = 'Commodity Code'
    fvalue = ET.SubElement(feature, 'FVALUE')
    fvalue.text = '{}'.format(i_key['FVALUE_CommodityCode'])


    # Article Order / Price / MIME
    article_order_details = ET.SubElement(article, 'ARTICLE_ORDER_DETAILS')

    order_unit = ET.SubElement(article_order_details, 'ORDER_UNIT')
    order_unit.text = '{}'.format(i_key['ORDER_UNIT'])

    content_unit = ET.SubElement(article_order_details, 'CONTENT_UNIT')
    content_unit.text = '{}'.format(i_key['CONTENT_UNIT'])

    no_cu_per_u = ET.SubElement(article_order_details, 'NO_CU_PER_OU')
    no_cu_per_u.text = '{}'.format(i_key['Menge pro Verkaufseinheit/ NO_CU_PER_OU'])

    price_quantity = ET.SubElement(article_order_details, 'PRICE_QUANTITY')
    price_quantity.text = '1'

    quantity_min = ET.SubElement(article_order_details, 'QUANTITY_MIN')
    quantity_min.text = '1'

    
    # PREIS for 1 qty
    article_price_details = ET.SubElement(article, 'ARTICLE_PRICE_DETAILS')

    article_price = ET.SubElement(article_price_details, 'ARTICLE_PRICE')
    article_price.set('price_type', "net_customer")

    price_amount = ET.SubElement(article_price, 'PRICE_AMOUNT')
    price_amount.text = '{}'.format(i_key['PREIS'])

    currency = ET.SubElement(article_price, 'PRICE_CURRENCY')
    currency.text = 'EUR'

    tax = ET.SubElement(article_price, 'TAX')
    tax.text = '0.19'

    lower_bound = ET.SubElement(article_price, 'LOWER_BOUND')
    lower_bound.text = '1'
    

    mime_info = ET.SubElement(article, 'MIME_INFO')

    #MIME 1
    mime = ET.SubElement(mime_info, 'MIME')
    mime_type = ET.SubElement(mime, 'MIME_TYPE')
    mime_type.text = 'image/jpeg'

    mime_source = ET.SubElement(mime, 'MIME_SOURCE')
    mime_source.text = '{}'.format(i_key['MIME_SOURCE1_lrg'])

    mime_descr = ET.SubElement(mime, 'MIME_DESCR')
    mime_descr.text = 'Thumbnail'

    mime_purpose = ET.SubElement(mime, 'MIME_PURPOSE')
    mime_purpose.text = 'normal'

    # MIME 2
    mime = ET.SubElement(mime_info, 'MIME')
    mime_type = ET.SubElement(mime, 'MIME_TYPE')
    mime_type.text = 'url'

    mime_source = ET.SubElement(mime, 'MIME_SOURCE')
    mime_source.text = '{}'.format(i_key['MIME_SOURCE1_lrg_weblink'])

    mime_descr = ET.SubElement(mime, 'MIME_DESCR')
    mime_descr.text = 'Web URL'

    mime_purpose = ET.SubElement(mime, 'MIME_PURPOSE')
    mime_purpose.text = 'detail'
```

```python
def display_elements(element, indent=''):
    print(f"{indent}Tag: {element.tag}, Attrib: {element.attrib}, Text: {element.text}")

    # Recursively iterate over the child elements
    for child in element:
        display_elements(child, indent + '  ')

```
